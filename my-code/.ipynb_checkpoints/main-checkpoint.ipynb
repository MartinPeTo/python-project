```python
# define rooms and items

#ROOMS

extrange_room = {
    "name": "extrange room",
    "type": "room",
}

big_bedroom = {
    "name": "big bedroom",
    "type": "room",
}

small_bedroom = {
    "name": "small bedroom",
    "type": "room",
}

living_room = {
    "name": "living room",
    "type": "room"
}

freedom = {
    "name": "freedom",
    "type": "room"
}

#DOORS

green_door = {
    "name": "green door",
    "type": "door",
    "lock": "locked"
}

red_door = {
    "name": "red door",
    "type": "door",
    "lock": "locked"
}

blue_door = {
    "name": "blue door",
    "type": "door",
    "lock": "locked"
}

black_door = {
    "name": "black door",
    "type": "door"
    "lock": "locked"
}

#FURNITURES

couch = {
    "name": "couch",
    "type": "furniture",
    "explore": "no explored"
}

piano = {
    "name": "piano",
    "type": "furniture",
    "explore": "no explored"
}

queen_bed = {
    "name": "queen bed",
    "type": "furniture",
    "explore": "no explored"
}

double_bed = {
    "name": "double bed",
    "type": "furniture",
    "explore": "no explored"
}

dresser = {
    "name": "dresser",
    "type": "furniture",
    "explore": "no explored"
}

dinning_table = {
    "name": "dinning table",
    "type": "furniture",
    "explore": "no explored"
}

#KEYS

green_key = {
    "name": "key for green door",
    "type": "key",
    "target": green_door,
}

blue_key = {
    "name": "key for blue door",
    "type": "key",
    "target": blue_door,
}

black_key = {
    "name": "key for black door",
    "type": "key",
    "target": black_door,
}

red_key = {
    "name": "key for red door",
    "type": "key",
    "target": red_door,
}

all_rooms = [extrange_room, big_bedroom, small_bedroom, living_room, freedom]

all_doors = [green_door, blue_door, red_door, black_door]

# define which items/rooms are related

object_relations = {
    "piano": [green_key],
    "queen bed": [red_key],
    "double bed": [blue_key],
    "dresser": [black_key],
}

room_relations = {
    "extrange room": [couch, piano, green_door],
    "big bedroom": [queen_bed, green_door, red_door, blue_door],
    "small bedroom": [double_bed, dresser, red_door],
    "living room": [dinning_table, black_door, blue_door],
}

door_relations = {
    "green door": [extrange_room, big_bedroom],
    "blue door": [big_bedroom, living_room],
    "red door": [big_bedroom, small_bedroom],
    "black door": [living_room, freedom]
}

# define game state. Do not directly change this dict. 
# Instead, when a new game starts, make a copy of this
# dict and use the copy to store gameplay state. This 
# way you can replay the game multiple times.

INIT_GAME_STATE = {
    "current_room": extrange_room,
    "keys_collected": [],
    "target_room": freedom
}
```

```python
def linebreak():
    """
    Print a line break
    """
    print("\n\n")

def start_game():
    """
    Start the game
    """
    print("You wake up on a couch and find yourself in a strange house with no windows which you have never been to before. You don't remember why you are here and what had happened before. You feel some unknown danger is approaching and you must get out of the house, NOW!")
    play_room(game_state["current_room"])

def play_room(room):
    """
    Play a room. First check if the room being played is the target room.
    If it is, the game will end with success. Otherwise, let player either 
    explore (list all items in this room) or examine an item found here.
    """
    game_state["current_room"] = room
    if(game_state["current_room"] == game_state["target_room"]):
        print("Congrats! You escaped the room!")
    else:
        print("You are now in a " + room["name"])
        intended_action = input("""What would you like to do? Type:
            'explore' to explore the room
            'examine' to examine an object
            'cross' to cross a door
            """).strip()
        if intended_action == "explore":
            explore_room(room)
            play_room(room)
        elif intended_action == "examine":
            examine_item(input("What would you like to examine? ").strip())
        elif intended_action == "cross":
            cross_door(input("Which door do you want to cross? ").strip())
        else:
            print("Not sure what you mean. Type 'explore', 'examine' or 'cross'.")
            play_room(room)
        linebreak()

def explore_room(room):
    """
    Explore a room. List all items belonging to this room.
    """
    items = [i["name"] for i in room_relations[room["name"]]]
    print("You explore the room. This is a " + room["name"] + ". There is a " + ", a ".join(items))

def get_next_room_of_door(door, current_room):
    """
    From object_relations, find the two rooms connected to the given door.
    Return the room that is not the current_room.
    """
    connected_rooms = door_relations[door["name"]]
    for room in connected_rooms:
        if(not current_room == room):
            return room

def examine_item(item_name):
    linebreak()

    current_room = game_state["current_room"]
    output = None
    
    for item in room_relations[current_room["name"]]:
        if item["name"] == item_name and item["type"] == "furniture":
            output = "You examine the " + item_name + "."
            if(item["name"] in object_relations):
                item_found = object_relations[item["name"]].pop()
                game_state["keys_collected"].append(item_found)
                output += "You find a " + item_found["name"] + "."
                linebreak()
            else:
                output += "Is just a regular " + item["name"] + "."
                linebreak()
                
            print(output)
            play_room(current_room)
            break

    if(output is None):
        print("In this room there isn't a " + item_name + ".")
        linebreak()
        play_room(current_room)

def cross_door(door_name):
    linebreak()
    
    current_room = game_state["current_room"]
    next_room = ""
    output = None
    for door in room_relations[current_room["name"]]:
        if(door["name"] == door_name and door["type"] == "door" and door["lock"] == "locked"):
            output = "You try to open the " + door_name + "."
            have_key = False
            for key in game_state["keys_collected"]:
                if(key["target"] == door):
                    have_key = True
            if(have_key):
                print('ok')
                output += "You unlock it with a key you have."
                next_room = get_next_room_of_door(door, current_room)
            else:
                output += "It's locked but you don't have the correct key."
        elif(door["name"] == door_name and door["type"] == "door" and door["lock"] == "unlocked"):
            next_room = get_next_room_of_door(door, current_room)
            output = "You open the " + door_name + "."
            
#ACABAR Y PROBAR DOBLE CHECK**********************************************************************

    print(output)
    linebreak()    
    if(output is None):
        print("In this room there isn't a" + item_name + ".")
        linebreak()
        play_room(current_room)
               
    if(next_room and input("Do you want to go to the next room? Enter 'yes' or 'no'").strip() == 'yes'):
        linebreak()
        play_room(next_room)
    else:
        linebreak()
        play_room(current_room)
```
